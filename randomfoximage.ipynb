```python
import requests
from IPython.display import Image, display

# RandomFox API URL
API_URL = "https://randomfox.ca/floof/"

def show_random_fox():
    # Get random fox data
    response = requests.get(API_URL)
    data = response.json()

    # Extract image URL
    fox_image_url = data["image"]

    print("Random Fox Image URL:")
    print(fox_image_url)
    print("\nDisplaying Fox Image:\n")

    # Display image in Colab output
    display(Image(url=fox_image_url))

# Generate and print random fox image
show_random_fox()
```
